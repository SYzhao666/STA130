In a simple linear regression model, we predict an outcome variable Y from a predictor X using the equation:
    Y = B0 + B1X + œµ

where: 
    B0 is the intercept (the value of Y when X = 0),
    B1 is the slope (the expected change in Y for a one-unit increase in X),
    œµ is the error term, which accounts for the deviation of actual values from the predicted line.

In this model, Y values are distributed normally around the line ùõΩ0+ùõΩ1ùëã with variance œÉ^2.
For each X, Y follows a normal distribution:
    Y‚àºN(Œ≤0 + Œ≤1X, œÉ^2)
In Python, we can simulate this model by generating X, computing the predicted Y, 
and adding normally distributed noise with standard deviation œÉ. 
The plot then shows data points scattered around the regression line ùê∏[ùëå‚à£ùëã]=ùõΩ0+ùõΩ1ùëã, 
demonstrating the linear relationship with normally distributed deviations.

````python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt

# Set random seed for reproducibility
np.random.seed(42)

# Parameters for the linear regression model
beta_0 = 5  # Intercept
beta_1 = 2  # Slope
sigma = 1   # Standard deviation of the error term

# Generate random samples for the predictor variable X
X = np.random.uniform(0, 10, 100)

# Calculate the mean of Y given X, which is the line of best fit without noise
mean_Y_given_X = beta_0 + beta_1 * X

# Generate the outcome variable Y by adding normally distributed noise to the mean
Y = mean_Y_given_X + norm.rvs(0, sigma, size=X.size)

# Plot the data
plt.scatter(X, Y, color="blue", alpha=0.6, label="Observed data (Y)")
plt.plot(X, mean_Y_given_X, color="red", label="True regression line (E[Y|X])")
plt.xlabel("Predictor (X)")
plt.ylabel("Outcome (Y)")
plt.title("Simple Linear Regression Model")
plt.legend()
plt.show()
````

````python
#link to ChatGPT transcript log: https://chatgpt.com/share/672bc172-3b80-800b-9e1f-8ed02d78e954
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.formula.api as smf

# Set random seed for reproducibility
np.random.seed(42)

# Parameters for the theoretical model
beta_0 = 5  # Intercept
beta_1 = 2  # Slope
sigma = 1   # Standard deviation of the error term

# Generate predictor variable X
X = np.random.uniform(0, 10, 100)

# Calculate outcome variable Y using the theoretical model
mean_Y_given_X = beta_0 + beta_1 * X
Y = mean_Y_given_X + np.random.normal(0, sigma, size=X.size)

# Create a DataFrame for the simulated data
data = pd.DataFrame({'X': X, 'Y': Y})

# Fit a simple linear regression model using statsmodels
model = smf.ols('Y ~ X', data=data).fit()

# Display the summary of the regression model
print(model.summary())

# Visualize the observed data and fitted regression line
plt.scatter(data['X'], data['Y'], color="blue", alpha=0.6, label="Observed data (Y)")
plt.plot(data['X'], model.fittedvalues, color="red", label="Fitted regression line")
plt.xlabel("Predictor (X)")
plt.ylabel("Outcome (Y)")
plt.title("Fitted Simple Linear Regression Model")
plt.legend()
plt.show()
````

````python
The fitted line chosen by the Ordinary Least Squares (OLS) method is the one that minimizes the sum of squared differences 
between the observed Y values and the predicted Y values from the regression model. 
In other words, it minimizes the sum of squared residuals (errors), 
where each residual is the vertical distance between an observed point and the line.

The reason it requires "squares" is to:
1.Eliminate negative values: Squaring the differences ensures that positive and negative errors don't cancel each other out.
2.Emphasize larger errors: Squaring the residuals gives more weight to larger deviations, 
    making the model sensitive to outliers or larger discrepancies.

Thus, the OLS method fits the line that best approximates the relationship between the predictor 
and outcome variables by minimizing these squared errors.
````

````python
#link to ChatGPT transcript log: https://chatgpt.com/share/672be49d-8bc8-800b-b250-f348c2b18fe8

#(A) Permutation Test for ùêª0:ùúá(short)=ùúá(long)

#Approach Explanation:
#A permutation test assesses whether there‚Äôs a significant difference between the mean durations of the ‚Äúshort‚Äù 
#and ‚Äúlong‚Äù wait times by randomly reassigning the group labels. 
#By calculating the difference in group means for each permutation, 
#we build a sampling distribution of the test statistic (mean difference) under the null hypothesis 
#that there's no actual difference between groups.

#Code for Permutation Test:

import numpy as np

# Observed mean difference
observed_difference = old_faithful.groupby('kind')['duration'].mean().iloc[::-1].diff().values[1]

# Permutation test setup
n_permutations = 1000
permuted_diffs = []

for _ in range(n_permutations):
    shuffled_data = old_faithful.assign(kind_shuffled=old_faithful['kind'].sample(frac=1, replace=False).values)
    permuted_diff = shuffled_data.groupby('kind_shuffled')['duration'].mean().iloc[::-1].diff().values[1]
    permuted_diffs.append(permuted_diff)

# Calculate p-value
p_value = (np.sum(np.abs(permuted_diffs) >= np.abs(observed_difference)) + 1) / (n_permutations + 1)

#This code calculates the mean difference for each permutation and finds a p-value 
#by comparing the observed mean difference to the distribution generated by shuffling.


#(B) Bootstrap Confidence Interval for Difference in Means
#Approach Explanation:
#The bootstrap method repeatedly samples (with replacement) within each group to generate bootstrapped samples. 
#For each sample, we calculate the difference in group means. This process builds a distribution of mean differences, 
#from which we can derive a 95% confidence interval.

#Code for Bootstrap Confidence Interval
# Bootstrap setup
n_bootstraps = 1000
bootstrapped_mean_diffs = []

for _ in range(n_bootstraps):
    bootstrapped_data = old_faithful.groupby('kind').apply(lambda x: x.sample(frac=1, replace=True)).reset_index(drop=True)
    boot_mean_diff = bootstrapped_data.groupby('kind')['duration'].mean().iloc[::-1].diff().values[1]
    bootstrapped_mean_diffs.append(boot_mean_diff)

# 95% confidence interval for the mean difference
ci_lower, ci_upper = np.quantile(bootstrapped_mean_diffs, [0.025, 0.975])

#This code calculates the mean difference for each bootstrapped sample and then determines the 95% confidence interval 
#for these differences.
````

````python
(a) Explanation of Sampling Approaches
1.Permutation Test: 
    By reshuffling the ‚Äúkind‚Äù labels, we simulate a scenario where the group membership is arbitrary, 
    building a distribution of mean differences that would be expected if there were no actual difference 
    between ‚Äúshort‚Äù and ‚Äúlong‚Äù groups.
2.Bootstrap Confidence Interval: 
    By repeatedly resampling with replacement within each group, 
    we simulate the sampling variability in the observed mean difference. 
    This approach provides a range of plausible values for the mean difference, 
    helping to estimate the true difference between the ‚Äúshort‚Äù and ‚Äúlong‚Äù groups.

(b) Comparison with the Indicator Variable Model Approach
1.Similarity:
    All methods estimate the difference between the ‚Äúshort‚Äù and ‚Äúlong‚Äù wait times, 
    assessing whether this difference is significant.
    The indicator variable model (regression approach) with C(kind) assesses the effect of ‚Äúlong‚Äù waits on eruption duration, 
    akin to comparing group means.
2.Difference:
    a.The indicator variable model estimates the mean difference between groups in a regression framework, 
    treating the kind variable as a categorical predictor with a statistical test of its coefficient. 
    This approach relies on the assumptions of linear regression, including normally distributed errors.
    b.The permutation test is non-parametric and does not rely on distributional assumptions about the residuals, 
    making it suitable when the data does not satisfy normality or homoscedasticity.
    c.The bootstrap confidence interval provides a non-parametric estimate of the range of plausible values for the difference 
    in means, independently of the assumptions of regression. 
    It captures sampling variability and is robust even if the underlying data distribution isn‚Äôt normal.

Interpretation
1.Permutation Test: 
    If the p-value is low, it indicates that the observed difference in means is unlikely under the null hypothesis.
2.Bootstrap Confidence Interval: 
    If this interval does not contain 0, it suggests a statistically significant difference in durations between ‚Äúshort‚Äù 
    and ‚Äúlong‚Äù waits.
Together, these methods provide complementary evidence regarding the difference between groups, 
with the indicator model formalizing this difference within a regression framework and the permutation 
and bootstrap approaches offering distribution-independent alternatives.
````

> The first three assumptions associated with the **Simple Linear regression** model are that
> 
> - the $\epsilon_i$ **errors** (sometimes referred to as the **noise**) are **normally distributed**
> - the $\epsilon_i$ **errors** are **homoscedastic** (so their distributional variance $\sigma^2$ does not change as a function of $x_i$)
> - the linear form is [at least reasonably approximately] "true" (in the sense that the above two remain [at least reasonably approximately] "true") so that then behavior of the $Y_i$ **outcomes** are represented/determined on average by the **linear equation**)<br>
> 
>    and there are additional assumptions; but, a deeper reflection on these is "beyond the scope" of STA130; nonetheless, they are that<br><br>
> - the $x_i$ **predictor variable** is **measured without error**
> - and the $\epsilon_i$ **errors** are **statistically independent** (so their values do not depend on each other)
> - and the $\epsilon_i$ **errors** are **unbiased** relative to the **expected value** of **outcome** $E[Y_i|x_i]=\beta_0 + \beta_1x_i$ (which is equivalently stated by saying that the mean of the **error distribution** is $0$, or again equivalently, that the **expected value** of the **errors** $E[\epsilon_i] = 0$)
