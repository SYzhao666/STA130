````python
#link to ChatGPT transcript log: https://chatgpt.com/share/6734f0fa-8da0-800b-973f-2ac88d150156
1. **Simple vs. Multiple Linear Regression**:
   - **Simple Linear Regression** models the relationship between one independent variable (predictor) 
    and one dependent variable (outcome), assuming a linear relationship. For example, predicting height based on age.
   - **Multiple Linear Regression** extends this by modeling the relationship between more than one independent variable 
    and the dependent variable. It accounts for multiple predictors at once, helping better capture the complexity of real-world relationships (e.g., predicting house prices based on square footage, number of bedrooms, and location).
   - **Benefit of Multiple Linear Regression**: It improves the accuracy and generalizability of the model 
    by incorporating multiple factors, thus providing a more complete explanation of the variance in the dependent variable.

2. **Continuous vs. Indicator Variable in Simple Linear Regression**:
   - A **continuous variable** is one that can take any value within a range (e.g., age, temperature). It shows a linear 
    relationship with the dependent variable.
   - An **indicator variable** (also called a dummy variable) is categorical, typically binary (0 or 1), and represents 
    categories like gender (male = 0, female = 1). It models a shift in the dependent variable between the groups.
   - **Linear Forms**: 
     - In Simple Linear Regression with a continuous variable, the equation would be `y = β0 + β1 * X + ε`.
     - In Simple Linear Regression with an indicator variable, the equation would be `y = β0 + β1 * D + ε`, where `D` is the 
    dummy variable (0 or 1).

3. **Effect of Adding an Indicator Variable to Create a Multiple Linear Regression**:
   - When a continuous variable and an indicator variable are included in **Multiple Linear Regression**, the model now accounts
    for the effect of both predictors. The model might look like `y = β0 + β1 * X + β2 * D + ε`, where `X` is the continuous 
    variable and `D` is the indicator.
   - **Expected Nature of the Data**: This model captures how both the continuous predictor and the group differences (captured 
    by the indicator) influence the dependent variable. The coefficient of the indicator variable (`β2`) models the shift in the
    outcome for different categories (like group A vs. group B), while the coefficient for the continuous variable (`β1`) shows 
    the effect of the continuous predictor.

4. **Adding Interaction Between Continuous and Indicator Variables**:
   - In **Multiple Linear Regression**, an **interaction term** between a continuous and an indicator variable models how the 
    effect of the continuous variable changes across the different categories of the indicator. This interaction term allows 
    for different slopes for each category.
   - The model would look like `y = β0 + β1 * X + β2 * D + β3 * (X * D) + ε`, where `β3` represents the interaction effect.
   - **Effect on Model**: The model now captures not only the individual effects of the continuous and categorical variables 
    but also how they combine, i.e., how the relationship between `X` and `y` differs depending on the value of `D`.

5. **Multiple Linear Regression with Only Indicator Variables**:
   - When **Multiple Linear Regression** is based only on indicator variables derived from a non-binary categorical variable, 
    the model creates separate binary variables for each category (using **one-hot encoding** or similar methods). This leads 
    to a model like `y = β0 + β1 * D1 + β2 * D2 + ... + βk * Dk + ε`, where `D1`, `D2`, ... are binary variables representing 
    the different categories.
   - **Behavior of the Model**: The model estimates the effect of each category relative to a baseline category (usually where 
    all indicator variables are 0). Each coefficient (`β1`, `β2`, etc.) represents the difference in the dependent variable 
    between the corresponding category and the baseline.
````

````python
#link to ChatGPT transcript log: https://chatgpt.com/share/6734f593-0bf0-800b-900b-3e71213e7e6f
import pandas as pd
import statsmodels.formula.api as smf

# Example dataset
url = "https://raw.githubusercontent.com/KeithGalli/pandas/master/pokemon_data.csv"
data = pd.read_csv(url)

# Create a binary outcome
data['binary_outcome'] = (data['Type 1'] == 'Fire').astype(int)

# Create a logistic regression formula for the additive model
formula_additive = 'binary_outcome ~ Attack + Legendary + C(Generation)'
log_reg_additive = smf.logit(formula_additive, data=data).fit()
print(log_reg_additive.summary())

# Create a logistic regression formula for the interaction model
# Replace 'Gender' with 'Legendary' for the interaction term
formula_interaction = 'binary_outcome ~ Attack * Legendary + C(Generation)'
log_reg_interaction = smf.logit(formula_interaction, data=data).fit()
print(log_reg_interaction.summary())

# Visualization (for continuous predictor with additive specification):
import plotly.express as px
import numpy as np

# Create a prediction range for 'Attack'
attack_range = np.linspace(data['Attack'].min(), data['Attack'].max(), 100)

# Define the correct numeric level for Generation (e.g., Generation I -> 1)
generation_mapping = {
    'Generation I': 1,
    'Generation II': 2,
    'Generation III': 3,
    'Generation IV': 4,
    'Generation V': 5,
    'Generation VI': 6
}

# Predict probabilities using the additive model
predicted_probs = log_reg_additive.predict(pd.DataFrame({
    'Attack': attack_range,
    'Legendary': [0] * 100,
    'Generation': [generation_mapping['Generation I']] * 100  # Use the numeric value for 'Generation I'
}))

# Plot the predicted probabilities
fig = px.line(x=attack_range, y=predicted_probs, labels={'x': 'Attack', 'y': 'Predicted Probability'})
fig.show(renderer="png")

# Simulating Noise and Plotting:
# Simulate random noise for 'Attack'
noise = np.random.normal(0, 0.1, len(data))

# Add noise to the data
noisy_data = data.copy()
noisy_data['Attack'] += noise

# Predict probabilities with noisy data
predicted_probs_with_noise = log_reg_additive.predict(noisy_data)

# Plot noisy data and predictions
fig = px.scatter(noisy_data, x='Attack', y='binary_outcome', title='Logistic Regression with Noise')
fig.add_scatter(x=noisy_data['Attack'], y=predicted_probs_with_noise, mode='lines', name='Fitted Line')
fig.show(renderer="png")
````
